```python
import pandas as pd 
 
df=pd.read_csv('cardio_train.csv', sep = ';')
print (df)
```

```python
df
```

```python
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Plot
sns.jointplot(x='weight', y='height',data=df, c='red', kind='kde')
plt.title('Weight Vs. Height Plot')
plt.xlabel('Weight of patients')
plt.ylabel('Height of patients')
plt.show()
```

```python
df = df.dropna()
plt.scatter(x='ap_hi', y='ap_lo',data=df, c='red')
plt.title('Systolic blood pressure Vs. Diastolic blood pressure')
plt.xlabel('Systolic blood pressure')
plt.ylabel('Diastolic blood pressure')
plt.show() 


```

```python
sns.jointplot(x='ap_hi', y='ap_lo',data=df, kind='reg')
plt.title('Systolic blood pressure Vs. Diastolic blood pressure')
plt.xlabel('Systolic blood pressure')
plt.ylabel('Diastolic blood pressure')
plt.show()
```
